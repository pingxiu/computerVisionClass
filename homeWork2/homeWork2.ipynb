```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('test.jpg')
img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
identity = np.array(([0,0,0], [0,1,0], [0,0,0]), dtype="int")
edge = np.array(([1,0,-1], [0,0,0], [-1,0,1]), dtype="int")
edge2 = np.array(([0, 1, 0], [1, -4, 1], [0, 1, 0]), dtype="int")
edge3 = np.array(([-1, -1, -1], [-1, 8, -1], [-1, -1, -1]), dtype="int")
sharp = np.array(([0, -1, 0],[-1, 5, -1], [0, -1, 0]), dtype="int")
box = np.ones((3, 3), dtype="float") / 9
gb = np.array(([1, 2, 1], [2, 4, 2], [1, 2, 1]), dtype="float") / 16
gb2 = np.array(([1, 4, 6, 4, 1], [4, 16, 24, 16, 4], [6, 24, 36, 24, 6], [4, 16, 24, 16, 4], [1, 4, 6, 4, 1]), dtype="float") / 256

dst = cv2.filter2D(img,-1,identity)
dst2 = cv2.filter2D(img,-1,edge)
dst3 = cv2.filter2D(img,-1,edge2)
dst4 = cv2.filter2D(img,-1,edge3)
dst5 = cv2.filter2D(img,-1,sharp)
dst6 = cv2.filter2D(img,-1,box)
dst7 = cv2.filter2D(img,-1,gb)
dst8 = cv2.filter2D(img,-1,gb2)

plt.imshow(dst)
plt.show()
plt.imshow(dst2)
plt.show()
plt.imshow(dst3)
plt.show()
plt.imshow(dst4)
plt.show()
plt.imshow(dst5)
plt.show()
plt.imshow(dst6)
plt.show()
plt.imshow(dst7)
plt.show()
plt.imshow(dst8)
plt.show()
```
