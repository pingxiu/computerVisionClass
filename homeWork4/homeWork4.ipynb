```python
import cv2
import numpy as np
import matplotlib.pyplot as pl
```

```python
# Read source image.
im_src = cv2.imread('dsu1.jpg')
# Four corners of the book in source image
pts_src = np.array([[324, 281], [326, 538] ,[760, 508],[768, 318]])
# Four corners of the book in destination image.
pts_dst = np.array([[320, 280], [320, 500] ,[760, 500],[760, 280]])
# Calculate Homography
h, status = cv2.findHomography(pts_src, pts_dst)
# Warp source image to destination based on homography
im_out = cv2.warpPerspective(im_src, h, (im_dst.shape[1],im_dst.shape[0]))
# Display images
pl.imshow(cv2.cvtColor(im_src, cv2.COLOR_BGR2RGB))
pl.show()
pl.imshow(cv2.cvtColor(im_out, cv2.COLOR_BGR2RGB))
pl.show()
```
